```python
import numpy as np

labels = ["Cat", "Dog", "Rabbit"]

# rows = predicted, cols = gold
cm = np.array([
    [5, 10, 5],
    [15, 20, 10],
    [0, 15, 10]
], dtype=float)

row_sums = cm.sum(axis=1)   # predicted totals
col_sums = cm.sum(axis=0)   # gold totals
tp = np.diag(cm)

precision = tp / row_sums
recall = tp / col_sums

macro_p = precision.mean()
macro_r = recall.mean()

micro_tp = tp.sum()
total = cm.sum()
micro_p = micro_tp / total
micro_r = micro_tp / total  # same for single-label multiclass

print("Per-class Precision/Recall:")
for i, lab in enumerate(labels):
    print(f"{lab:6s}  Precision={precision[i]:.4f}  Recall={recall[i]:.4f}")

print("\nMacro-averaged:")
print(f"Precision={macro_p:.4f}  Recall={macro_r:.4f}")

print("\nMicro-averaged:")
print(f"Precision={micro_p:.4f}  Recall={micro_r:.4f}")
```
